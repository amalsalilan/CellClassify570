```python
from imblearn.over_sampling import ADASYN
```

 the 'id' column and the 'Unnamed: 32' column (which appears to be empty) from your dataset:

```python
df = df.drop(['id', 'Unnamed: 32'], axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['diagnosis']=le.fit_transform(df['diagnosis'])
```

```python
df
```

```python
x=df.drop(columns='diagnosis')
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
adasyn=ADASYN(random_state=42)
X_train_balnced,y_train_balnced=adasyn.fit_resample(x_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,recall_score,f1_score
models = [
    ('Logistic Regression', LogisticRegression()),
    ('Decision Tree', DecisionTreeClassifier()),
    ('Random Forest', RandomForestClassifier()),
    ('Support Vector Machine', SVC())
]

for name, model in models:
    model.fit(X_train_balnced, y_train_balnced)
    y_pred = model.predict(x_test)
    accuracy = recall_score(y_test, y_pred)
    print(f" Recall  :{name}: {accuracy:.4f}")
    f1=f1_score(y_test,y_pred)
    print(f"f1 score :{name}:{f1:.4f}")
```

In the given output, we can see that the Decision Tree, Random Forest, and Support Vector Machine models achieve high recall scores, with SVM achieving the highest recall score. Therefore, SVM could be a good choice
